Loan Repayment or not using DecisionTree

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
balance_data = pd.read_csv(r'C:\Users\PC\Downloads\Decision_Tree_ Dataset.csv',
sep= ',', header= 0)
balance_data
```

```python
balance_data.head()
```

```python
balance_data.rename(columns={'1': 'initial payment', '2': 'last payment','3':'Credit score','4':'House Number','Unnamed: 5':'Result'}, inplace=True)
balance_data
```

```python
balance_data.drop(columns=['sum'], inplace=True)
balance_data
```

```python
X = balance_data.values[:, 0:4]
Y = balance_data.values[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)
```

```python
#function to perform training with entropy
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
 max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred_en = clf_entropy.predict(X_test)
y_pred_en
```

```python
#checking accuracy
print ("Accuracy is "), accuracy_score(y_test,y_pred_en)*100
```
